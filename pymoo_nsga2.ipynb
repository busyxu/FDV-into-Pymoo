```python
# -- coding: utf-8 --
"""
@Project: pymoo
@Time : 2022/1/27 15:25
@Author : Yang xu
@Site : 
@File : test1.py
@IDE: PyCharm
"""
from pymoo.algorithms.moo.nsga2 import NSGA2
from pymoo.algorithms.moo.fdvga import FDVGA
from pymoo.factory import get_problem, get_sampling, get_crossover, get_mutation, get_selection
from pymoo.optimize import minimize
from pymoo.visualization.scatter import Scatter
```

# 测试问题：DTLZ1 ，目标数：2，决策变量：500，进化代数：2500

```python
problem = get_problem("dtlz1",n_var=500, n_obj=2)
```

```python
algorithm = NSGA2(pop_size=100)
```

```python
res = minimize(problem,
               algorithm,
               ('n_gen', 2500),
               seed=1,
               save_history=True,
               verbose=True)

X = res.X
F = res.F
```

```python
print("测试问题：DTLZ1 ，目标数：2，决策变量：500，进化代数：2500")
plot = Scatter()
plot.add(res.F, color="red")
plot.show()
```
