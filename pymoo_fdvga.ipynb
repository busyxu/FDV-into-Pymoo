## 测试问题：DTLZ1 ，目标数：2，决策变量：500，进化代数：2500

```python
2500*0.6
```

```python
problem = get_problem("dtlz1", n_var=500, n_obj=2)
```

```python
# algorithm = FDVGA(problem=problem, rate=0.8, acc=0.4, pop_size=10)
algorithm = FDVGA(rate=0.8, acc=0.4, max_gen=2500, pop_size=100)
# algorithm = NSGA2(pop_size=100,
#                   sampling=get_sampling("bin_random"),
#                   selection=get_selection('random'),
#                   crossover=get_crossover("bin_two_point"),
#                   mutation=get_mutation("bin_bitflip"),
#                   eliminate_duplicates=True)

# algorithm = NSGA2(pop_size=50)
```

```python

# res = minimize(problem,
#                algorithm,
#                ('n_gen', 200),
#                seed=1,
#                verbose=False)
res = minimize(problem,
               algorithm,
               ('n_gen', 2500),
               seed=1,
               save_history=True,
               verbose=True)

X = res.X
F = res.F
```

```python
print("测试问题：DTLZ1 ，目标数：2，决策变量：500，进化代数：2500")
plot = Scatter()
plot.add(res.F, color="red")
plot.show()
```
